```python
import seaborn as sns

sns.set()

sns.swarmplot(x='City', y='Rating', data=MyDataset)
plt.xlabel('City')
```

```python
sns.regplot(x='Tax 5%',y='gross income',data=MyDataset,color='Red')
plt.xlabel('Pajak 5%')
plt.ylabel('Pendapatan Kotor')
plt.show()
```

```python
sns.distplot(MyDataset['Rating'],kde=False)
plt.show()
```

```python
MyDataset.hist(figsize=(12,12))
plt.show()
```

```python
sns.countplot(MyDataset['Branch'])
plt.xlabel('Nama Branch')
plt.ylabel('Frekuensi')
plt.title('Branch yang paling banyak dikunjungi')
A,B,C =MyDataset.Branch.value_counts()
plt.savefig('report_Branch.png')

print('Branch A -',A)
print('Branch B -',C)
print('Branch C -',B)
      

plt.show()
```

```python
sns.countplot(MyDataset['Payment'])
plt.xlabel('Cara Pembayaran')
plt.ylabel('Frekuensi')
plt.title('Diagram Metode Pembayaran yang sering digunakan')
A,B,C =MyDataset.Payment.value_counts()
plt.savefig('report_Payment.png')

print('E-wallet -',A)
print('Cash -',B)
print('Credit Card -',C)
plt.show()
```

```python
sns.countplot(MyDataset['Gender'])
plt.xlabel('Gender')
plt.ylabel('Frekuensi')
plt.title('Frekuensi Paling sering berbelanja berdasarkan Jenis Kelamin')
A,B =MyDataset.Gender.value_counts()
plt.savefig('report_Gender.png')

print('Laki-laki -',B)
print('Perempuan -',A)

plt.show()
```

```python
sns.countplot(MyDataset['Customer type'])
plt.xlabel('Level Customer')
plt.ylabel('Frekuensi')
plt.title('Frekuensi pembelian tiap customer berdasarkan Level')
plt.savefig('report_Type.png')

plt.show()
```

```python
sns.countplot(y ='Product line', hue = "Gender", data = MyDataset) 
plt.xlabel('Frekuensi')
plt.ylabel('Tipe Produk')
plt.savefig('report_ProdukGender.png')
plt.show()
```
